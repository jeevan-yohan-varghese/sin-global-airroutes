```python
df_routes = pd.read_csv("routes.csv",names=['Sl', 'FNO','Dep','DepId' ,'Arr','ArrId','X','Y','Z'])
df_airports = pd.read_csv("airports.csv")
df_routes.head()
```

```python

df_routes['DepId']=df_routes['DepId'].replace('\\N', np.nan)
df_routes['ArrId']=df_routes['ArrId'].replace('\\N', np.nan)
df_routes=df_routes.dropna()
df_routes['DepId']=df_routes['DepId'].replace('\\N', np.nan).astype(str).astype(int)
df_routes['ArrId']=df_routes['ArrId'].replace('\\N', np.nan).astype(str).astype(int)
df_merged = pd.merge(df_routes, df_airports, left_on='DepId', right_on='id')
df_merged = pd.merge(df_merged, df_airports, left_on='ArrId', right_on='id', suffixes=('_Dep', '_Arr'))
#merged_df.head()
print(df_merged.columns)
df_merged.rename(columns = {'latitude_deg_Arr':'arr_lat','longitude_deg_Arr':'arr_lon','longitude_deg_Dep':'dep_lon',
                            'latitude_deg_Dep':'dep_lat'
                            }, inplace = True)
print(df_merged.columns)
```

```python
df_airports.head()
```

```python
fig = go.Figure()

fig.add_trace(go.Scattergeo(
    lon = df_merged['arr_lon'].append(df_merged['dep_lon']),
    lat = df_merged['arr_lat'].append(df_merged['dep_lat']),
    mode = 'markers',
    marker = dict(
    size = 4,
    color = 'rgb(255, 0, 0)',
    line = dict(width = 3,color = 'rgba(68, 68, 68, 0)'),
    )))

for i in range(len(df_merged)):
    fig.add_trace(
        go.Scattergeo(
            lon = [df_merged['arr_lon'][i], df_merged['dep_lon'][i]],
            lat = [df_merged['arr_lat'][i], df_merged['dep_lat'][i]],
            mode = 'lines',
            line = dict(width = 1.5,color = 'red'),
            opacity=0.3
            # opacity = float(df_merged['nb_flights'][i]) / float(df_merged['nb_flights'].max()),
        )
    )
fig.show()
```
